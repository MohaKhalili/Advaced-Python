```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.random(400)
y1 = np.random.random(400)

x2 = -1*np.random.random(400)
y2 = np.random.random(400)

x3 = np.random.random(400)
y3 = -1*np.random.random(400)

x4 = -1*np.random.random(400)
y4 = -1*np.random.random(400)

x = np.zeros(1600)
q = np.zeros(1600)
for i in range(1600):
  if i<400:
    x[i] = x1[i]
    q[i] = y1[i]
  elif i<800:
    x[i] = x2[i-400]
    q[i] = y2[i-400]
  elif i<1200:
    x[i] = x3[i-800]
    q[i] = y3[i-800]
  elif i<1600:
    x[i] = x4[i-1200]
    q[i] = y4[i-1200]

y = np.zeros(1600)
y[000:400] = 0
y[400:800] = 1
y[800:1200] = 1
y[1200:1600] = 0

x.shape
q.shape
y.shape

plt.scatter(x,q,c=y);
```
